
### Завдання 4: Транспонування матриць
1. Транспонуйте `result_matrix`.
2. Зробіть транспонування вектора `vector` і виведіть форму нової структури. Чи бачите ви зміну?
3. Використайте операцію `numpy.expand_dims` над вектором `vector` з аргументом axis=1. Запишіть отриманий результат в змінну `column_vector` і виведіть на екран. Порівняйте з `vector`.
4. Зробіть транспонування `column_vector`, виведіть результат на екран разом з розмірностями отриманої структури.




```python
vector = np.array([0.18, 0.53, 0.75, 0.47, 0.35, 0.21, 0.17, 0.58, 0.75, 0.38])

A = np.array([
    [0.18, 0.53],
    [0.75, 0.47],
    [0.35, 0.21]
])
B = np.array([
    [0.17, 0.58, 0.75],
    [0.38, 0.11, 0.15]
])

result_matrix = np.dot(A, B)  # 3x3
print("result_matrix:\n", result_matrix)
print("Shape:", result_matrix.shape)

result_matrix_T = result_matrix.T
print("\nTransposed result_matrix:\n", result_matrix_T)
print("Shape:", result_matrix_T.shape)

vector_T = vector.T
print("\nOriginal vector:\n", vector)
print("Shape:", vector.shape)
print("Transposed vector (1D):\n", vector_T)
print("Shape:", vector_T.shape)

column_vector = np.expand_dims(vector, axis=1)
print("\nColumn vector:\n", column_vector)
print("Shape:", column_vector.shape)

column_vector_T = column_vector.T
print("\nTransposed column_vector:\n", column_vector_T)
print("Shape:", column_vector_T.shape)
```
