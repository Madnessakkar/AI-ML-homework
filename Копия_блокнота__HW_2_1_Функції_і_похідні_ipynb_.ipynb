1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

f = 10 * x - 5

df = np.full_like(x, 10)

plt.figure()
plt.plot(x, f, label="f(x) = 10x - 5")
plt.plot(x, df, label="f'(x) = 10")
plt.xlabel("Кількість годин навчання")
plt.ylabel("Оцінка / Похідна")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Похідна функції дорівнює 10
#Це означає, що:
#Зі збільшенням кількості годин навчання оцінка зростає лінійно, і кожна додаткова година навчання підвищує оцінку в середньому на 10 балів.
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.symbols('x')

f1 = sp.sin(x)
f2 = sp.log(x)

df1 = sp.diff(f1, x)
df2 = sp.diff(f2, x)

print("f(x) = sin(x), f'(x) =", df1)
print("f(x) = ln(x), f'(x) =", df2)
```

```python
x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)

f_vals = np.sin(x_vals)
df_vals = np.cos(x_vals)

plt.figure()
plt.plot(x_vals, f_vals, label="f(x) = sin(x)")
plt.plot(x_vals, df_vals, label="f'(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_vals = np.linspace(0.1, 5, 400)

f_vals = np.log(x_vals)
df_vals = 1 / x_vals

plt.figure()
plt.plot(x_vals, f_vals, label="f(x) = ln(x)")
plt.plot(x_vals, df_vals, label="f'(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5*x + 1

df = sp.diff(f, x)

print("f(x) =", f)
print("f'(x) =", df)
```

```python
x_vals = np.linspace(0, 100, 400)

f_vals = x_vals**2 - 5*x_vals + 1
df_vals = 2*x_vals - 5

plt.figure()
plt.plot(x_vals, f_vals, label="f(x) = x² - 5x + 1")
plt.plot(x_vals, df_vals, label="f'(x) = 2x - 5")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Значення")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Функція витрат f(x) — квадратична, витрати зростають прискорено зі збільшенням обсягу виробництва
#Чим більше виробництво — тим швидше ростуть витрати
```
